```python
import torch
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
def set_all_seeds(seed):
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

# Set a fixed seed for reproducibility
SEED = 42
set_all_seeds(SEED)
```

```python
# Update train_test_split to use a fixed random_state
X_train, X_test, y_train , y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0], test_size=0.2, random_state=SEED)
```

#scaling


When using StandardScaler (or any other fit-based transformer), fit_transform is applied to X_train to calculate the mean and standard deviation from the training data and then scale it. This is crucial because it prevents 'data leakage' from the test set into the training process. The transform method is then applied to X_test using the same mean and standard deviation learned from X_train. This ensures that the test data is scaled consistently with the training data, without the test data influencing the scaling parameters, which could lead to an overly optimistic evaluation of the model's performance.


```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train[:5,:]
```

#Label Encoding


```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)

```

# Numpy arrays to Pytorch tensors

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
y_train_tensor_processed= y_train_tensor.double().unsqueeze(1)
```

#Defining the model


```python
# only one neuron
class MySimpleNN():
  def __init__(self,X):
    self.weights = torch.rand(X.shape[1],1, dtype = torch.float64, requires_grad=True)
    self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True) # 1 bias for 1 neuron

  def forward(self,X):
      self.output =  torch.matmul(X, self.weights) + self.bias
      return self.output

  def loss_function(self,y_pred,y_true):

      # clamp predictions to avoid log(0)

      epsilon = 1e-7
      y_pred = torch.clamp(y_pred, epsilon, 1-epsilon)
      #calculate loss : average
      loss=-(y_true * torch.log(y_pred) + (1 - y_true) * torch.log(1 - y_pred)).mean()
      return loss

class Activation_sigmoid():

  def forward(self, Z):
    self.output = torch.sigmoid(Z)
    return self.output



```

#Important Parameters

```python
learning_rate = 0.1
epochs = 25
```

# Training Pipeline


```python
model = MySimpleNN(X_train_tensor)
model.weights
model.bias

# define loop ( epoch )
for  epoch in range(epochs):


# forward pass

  layer_output = model.forward(X_train_tensor)

  sigmoid = Activation_sigmoid()
  y_pred = sigmoid.forward(layer_output)





# loss calculate
  loss = model.loss_function(y_pred,y_train_tensor)
  print(loss)







# backward pass

  loss.backward()

# # weight and bias update
  with torch.no_grad():
    model.weights -=  learning_rate * model.weights.grad
    model.bias-= learning_rate * model.bias.grad
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  print(f'Epoch: {epoch+1}, Loss: {loss.item()}')

```

# model evaluation

```python
with torch.no_grad():
  y_pred = model.forward(X_test_tensor)
  y_pred = sigmoid.forward(y_pred)
  y_pred = (y_pred > 0.9).float()
  accuracy = ( y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
