```python
# model with single neuron
# 5 features and 1 output

# create model class

import torch
import torch.nn as nn

class Model(nn.Module): # inherit from the nn.Module class
  def __init__(self,num_features):
    super().__init__()
    # create layers
    self.linear = nn.Linear(num_features,1) # no of inputs and no of outputs(no of neurons)
    # activation function
    self.sigmoid = nn.Sigmoid()

  # forward pass

  def forward(self,features):
    out = self.linear(features)
    out = self.sigmoid(out)
    return out


```

```python
# create dummy dataset

features = torch.rand(10,5) # 5 features

# create model instance

model = Model(features.shape[1])

# call model for forward pass

# model.forward(features) : this is not recommended\

model(features) # auto call forward ( overrides __call__ method)
```

```python
# we will create a model with 1 hidden layer and 1 output layer

# 5 input -> 3 neurons (15 weights and 3 bias) -> 1 output neuron ( 3* 1 weights) (1 bias)

# weight matrix size will be (3*5)

class Model_2(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    # hidden layer 1
    self.hidden_layer = nn.Linear(num_features,3) # 3 neurons
    self.relu = nn.ReLU()
    # output layer
    self.out_layer =  nn.Linear ( 3, 1)
    self.sigmoid = nn.Sigmoid()


  def forward(self, features):
    hidden_out = self.hidden_layer(features)
    hidden_out = self.relu(hidden_out)
    out = self.out_layer(hidden_out)
    out = self.sigmoid(out)
    return out


```

```python
model_2 =Model_2(features.shape[1])
print(model_2(features))

print(summary(model_2, input_size = (10,5)))
```

```python
# simplifying the forwards pass with nn.Sequential container

class Model_2(nn.Module):
  def __init__(self, num_features):
    super().__init__()

    self.network = nn.Sequential(
    nn.Linear(num_features,3), # 3 neurons
    nn.ReLU(),
    nn.Linear( 3, 1),
    nn.Sigmoid()
    )

  def forward(self, features):

    out = self.network(features)
    return out
```

```python
model_2 =Model_2(features.shape[1])
print(model_2(features))

print(summary(model_2, input_size = (10,5)))
```
