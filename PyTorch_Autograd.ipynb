#Autograd
Autograd  provides automatic differentiation for tensor operations. It enables gradient computation, which is essential for training machine learning models using optimization alogrihtms like gradient descent


```python
import torch
```

```python
x = torch.tensor(3.0,requires_grad=True)  #scalar tensor , bu default requires_grad= False
z= torch.tensor(2.0,requires_grad=True)
y= x**2 + z**4
```

```python
print(x,y) # grad_fn = <PowBackward0 )# internally computes computational graph x->sq->y: forward,  y->sq->x : backward ( grad )
```

```python
# run backward propagation from y

y.backward()
```

```python
# find gradient w.r.t x
x.grad
# find gradient w.r.t z
z.grad
```

```python
x = torch.tensor(3.0,requires_grad=True)
y=x**2
z= torch.sin(y)
z.backward()
x.grad
#y.grad can't compute this. y is internal nodes in the computational graph
```

```python
# for a perceptron
# x: cgpa , y: placement
x = torch.tensor(3.5 , requires_grad=True)
b= torch.tensor(1.0, requires_grad=True)
w = torch.rand(1, requires_grad=True)
z = w*x + b
print(z)
# activation function
y_pred = torch.sigmoid(z)
y_target = 1
loss= - (y_target*torch.log(y_pred) + (1-y_target)*torch.log(1-y_pred))
loss.backward()
print(w.grad)
print(b.grad)

w = w+ w.grad
b=b+b.grad

print(w,b)
```

```python
# clearing grad , # for multiple time backward operation gradient accumulates with new pass, so we will not get actual value
x.grad.zero_()
```
