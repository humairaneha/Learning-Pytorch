```python
#round
c= torch.tensor([1.9, 2.5, 2.3, 3.7,4.4])
print(torch.round(c))
print(torch.floor(c))
print(torch.ceil(c))
print(torch.clamp(c, min=2,max=3)) # fixed range
```

# Reduction Operation

```python
e= torch.randint(low=0,high=10,size=(2,3)) #This function generates a tensor filled with random integers.
print(e)
```

```python
#sum
print(torch.sum(e))
#sum along columns
print(torch.sum(e,dim=0)) #reducing row
#sum along rows
print(torch.sum(e,dim=1)) #reducing column
```

```python
#mean ( to calculate mean we need to have long type)
print(torch.mean(e,dtype=torch.float32))
#mean along col
print(torch.mean(e,dtype=torch.float32,dim=0))
#mean along row
print(torch.mean(e,dtype=torch.float32,dim=1))
```

```python
#meadian
torch.median(e)
#max and min
print(torch.max(e))
print(torch.min(e))
print(torch.argmax(e))
print(torch.argmin(e))
```

```python
#statistical operation
# standard deviation
torch.std(e.to(torch.float32))
torch.var(e.to(torch.float32))
```

# Matrix operations


```python
f = torch.tensor([[1,2,3],[4,5,6]] , dtype= torch.float32)
g = torch.tensor([[1,1],[2,2],[3,3]] , dtype= torch.float32)
```

```python
#matrix multiplication
k =torch.matmul(f,g)
```

```python
#scalar dot product
vec1 = torch.tensor([1,2,3])
vec2 = torch.tensor([4,5,6])
torch.dot(vec1,vec2)
```

```python
#transpose
torch.manual_seed(42)
x=torch.randn(size=(3,3))
print(x)
torch.transpose(x,0,1)
```

```python
#determinant
torch.det(k)
#inverse matrix
p = torch.tensor([[1,2],[3,4]],dtype=torch.float32)
torch.inverse(p)
```

```python
#special functions
torch.manual_seed(42)

func = torch.randn(2,3)
print(func)

# log

print( torch.log(func) )

#exp

print(torch.exp(func))

#sqrt

print(torch.sqrt(func))

#sigmoid

print(torch.sigmoid(func))

#softmax
print(torch.softmax(func,dim=1))

#relu
print(torch.relu(func))
```

#Tensor operations on GPU


```python
torch.cuda.is_available()
```

```python
device = torch.device('cuda') # storing gpu device
```

```python
#creating a new tensor to gpu
torch.rand((2,3), device=device)
```

```python
#moving existing tensor to GPU
a = torch.rand(2,3) # on cpu
b = a.to(device)
b+5
```

```python
import time

size = 10000

# tensors on CPU
mat_cpu1 = torch.randn(size,size)
mat_cpu2 = torch.randn(size,size)

# measure time on CPU
start_time = time.time()
result_cpu = torch.matmul(mat_cpu1,mat_cpu2)
cpu_time = time.time() - start_time
print(f"Time on CPU: {cpu_time:.4f} seconds")

mat_gpu1 = mat_cpu1.to(device)
mat_gpu2 = mat_cpu2.to(device)

# measure time on GPU
start_time = time.time()
result_gpu = torch.matmul(mat_gpu1,mat_gpu2)
torch.cuda.synchronize() # ensure all gpu operations are complete

gpu_time = time.time() - start_time
print(f"Time on GPU: {gpu_time:.4f} seconds")

print(cpu_time/gpu_time)
```
